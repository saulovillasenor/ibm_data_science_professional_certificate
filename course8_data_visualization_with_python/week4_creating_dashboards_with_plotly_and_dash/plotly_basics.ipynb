```python
# Import required libraries

import pandas as pd
import plotly.express as px
import plotly.graph_objects as go
```

# Read Data


```python
# Read the airline data into pandas dataframe
airline_data =  pd.read_csv('https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-DV0101EN-SkillsNetwork/Data%20Files/airline_data.csv', 
                            encoding = "ISO-8859-1",
                            dtype={'Div1Airport': str, 'Div1TailNum': str, 
                                   'Div2Airport': str, 'Div2TailNum': str})
```

```python
# Preview the first 5 lines of the loaded data 
airline_data.head()
```

```python
# Shape of the data
airline_data.shape
```

```python
# Randomly sample 500 data points. Setting the random state to be 42 so that we get same result.
data = airline_data.sample(n=500, random_state=42)
```

```python
# Get the shape of the trimmed data
data.shape
```

# plotly.graph_objects¶


## 1. Scatter Plot


Let's see how departure time changes with respect to airport distance

```python
# First we create a figure using go.Figure and adding trace to it through go.scatter
fig = go.Figure(data=go.Scatter(x=data['Distance'], y=data['DepTime'], mode='markers', marker=dict(color='red')))
# Updating layout through `update_layout`. Here we are adding title to the plot and providing title to x and y axis.
fig.update_layout(title='Distance vs Departure Time', xaxis_title='Distance', yaxis_title='DepTime')
# Display the figure
fig.show("svg")
```

## 2. Line Plot


Let's extract the average monthly arrival delay time and see how it changes over the year.


```python
# Group the data by Month and compute average over arrival delay time.
line_data = data.groupby('Month')['ArrDelay'].mean().reset_index()
```

```python
# Display the data
line_data
```

Create a line plot with x-axis being the month and y-axis being computed average delay time. Update plot title, xaxis, and yaxis title.



```python
fig = go.Figure(data=go.Scatter(x=line_data['Month'], y=line_data['ArrDelay'], mode='lines', marker=dict(color='green')))
fig.update_layout(title='Month vs Average Flight Delay Time', xaxis_title='Month', yaxis_title='ArrDelay')
fig.show("svg")
```

# plotly.express¶


## 1. Bar Chart


Let's extract the number of flights from a specific airline that goes to a destination


```python
# Group the data by destination state and reporting airline. Compute total number of flights in each combination
bar_data = data.groupby(['DestState'])['Flights'].sum().reset_index()
```

```python
# Display the data
bar_data
```

```python
# Use plotly express bar chart function px.bar. Provide input data, x and y axis variable, and title of the chart.
# This will give total number of flights to the destination state.
fig = px.bar(bar_data, x="DestState", y="Flights", title='Total number of flights to the destination state split by reporting airline') 
fig.show("svg")
```

## 2. Bubble Chart


Let's get number of flights as per reporting airline


```python
# Group the data by reporting airline and get number of flights
bub_data = data.groupby('Reporting_Airline')['Flights'].sum().reset_index()
```

```python
bub_data
```

Let's:

1. Create a bubble chart using the `bub_data` with x-axis being reporting airline and y-axis being flights.
2. Provide title to the chart
3. Update size of the bubble based on the number of flights. Use `size` parameter.
4. Update name of the hover tooltip to `reporting_airline` using `hover_name` parameter.


```python
fig = px.scatter(bub_data, x="Reporting_Airline", y="Flights", size="Flights",
                 hover_name="Reporting_Airline", title='Reporting Airline vs Number of Flights', size_max=60)
fig.show("svg")
```

# Histogram


Let's get the distribution of arrival delay


```python
# Set missing values to 0
data['ArrDelay'] = data['ArrDelay'].fillna(0)
```

Noe let us:

1. Use px.histogram and pass the dataset.
2. Pass `ArrDelay` to x parameter.


```python
fig = px.histogram(data, x="ArrDelay")
fig.show("svg")
```

# Pie Chart


Let's check the proportion of distance group by month (month indicated by numbers)


```python
# Use px.pie function to create the chart. Input dataset. 
# Values parameter will set values associated to the sector. 'Month' feature is passed to it.
# labels for the sector are passed to the `names` parameter.
fig = px.pie(data, values='Month', names='DistanceGroup', title='Distance group proportion by month')
fig.show("svg")
```

# Sunburst Charts


Here we will hierarchical view in the order of month and destination state holding value of number of flights


We will:

1. Create sunburst chart using `px.sunburst`.
2. Define hierarchy of sectors from root to leaves in `path` parameter. Here, we go from `Month` to `DestStateName` feature.
3. Set sector values in `values` paramter. Here, we can pass in `Flights` feature.
4. Show the figure.

```python
fig = px.sunburst(data, path=['Month', 'DestStateName'], values='Flights')
fig.show("svg")
```
